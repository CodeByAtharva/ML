```python
#this is the simple way but in the dataset weekend is a boolean values still it is giving us a mean
print(ds.mean(numeric_only=True))
print(ds.median(numeric_only=True))
print(ds.mode(numeric_only=True))
```

```python
#g) Find Mean,Median and Mode
n_ds=ds.select_dtypes(['int','float'])
print(n_ds.mean())
print(n_ds.median())
n_ds.mode()
```

```python
#h) Replace the missing values
ds=ds.fillna(ds.median(numeric_only=True))
ds.isnull().sum()
```

```python
#if we want to only replace the numeric column values not string and booleans
numeric_cols = ds.select_dtypes(include=['int64', 'float64']).columns #list of column names that has the numerical values and stored inside the var.
ds[numeric_cols] = ds[numeric_cols].fillna(ds[numeric_cols].median())
```

```python
n_col=ds.select_dtypes(['int','float']).columns
ds[n_col]=ds[n_col].fillna(ds[n_col].median())
```
